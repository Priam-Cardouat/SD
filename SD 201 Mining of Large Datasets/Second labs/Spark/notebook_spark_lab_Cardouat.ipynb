```
text_file=sc.textFile("/FileStore/tables/steve.txt")
words = text_file.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda s : ( s , 1) )
counts = pairs.reduceByKey( lambda a , b : a + b)
for (count,word) in counts.collect():
  print ("%s : %i" % ( count , word ))
```

```
text_file=sc.textFile("/FileStore/tables/steve.txt")
words = text_file.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda s : ( s , 1) )
counts = pairs.reduceByKey( lambda a , b : a + b)
counts=counts.map(lambda l: (l[1],l[0]))
counts=counts.sortByKey(False)
counts=counts.map(lambda l: (l[1],l[0]))
for (count,word) in counts.take(5):
  print ("%s : %i" % ( count , word ))
```

```
text_file1=sc.textFile("/FileStore/tables/idslabels.txt")
text_file2=sc.textFile("/FileStore/tables/edgelist.txt")

edges = text_file2.map(lambda line : line.split(" "))
edges = edges.flatMap(lambda x: x[1:]) #in order not to take the index number

pairs = edges.map(lambda s : (s,1))
counts = pairs.reduceByKey(lambda a, b :a+b)
counts = counts.map(lambda l : (l[1],l[0]))  
counts = counts.sortByKey(False)

names = text_file1.map(lambda line : line.split(" "))
pairs2 =names.map(lambda s : (int(s[0]),' '.join(s[1:])))

for (count,id) in counts.take(10):
  l = pairs2.filter(lambda s: s[0] ==int(id)).collect()
  print ("%s: %i" % (l[0][1],count))
```
